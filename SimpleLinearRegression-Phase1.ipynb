```python
#pandas is a python library. It contains the functions to load the data and process the same
#pandas is imported to load the input data from client into our program to develop the model
import pandas as pd
```

```python
#reading the input csv file using the read_csv function in pandas library
dataset=pd.read_csv("Salary_Data.csv")
dataset
```

```python
#indetify the input column and load the same to independent variable
independent=dataset[["YearsExperience"]]
independent
```

```python
#Identify the output/to predict column and load the same to dependent variable
dependent=dataset[["Salary"]]
dependent
```

```python
#sklearn is a python librarary. It contains the functions related to regression
#import the train_test_split function from the sklearn.model_selection package to split train and test dataset
from sklearn.model_selection import train_test_split
#train_test_split will get 4 inputs(independent, dependent, test_size,random_state)
#independent is a input dataset
#dependet is a output/to predict dataset
#test_size is a ratio of testdata
#random_state

#train_test_split will give 4 outputs(X_train, X_test, Y_train,Y_test) based on the test_size ratio
#X_train is a train dataset input
#Y_train is a train  output/to predict datset
#X_test is a test dataset input
#Y_test is a test  output/to predict datset
X_train, X_test, Y_train, Y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
print(X_train)
print(X_test)
print(Y_train)
print(Y_test)
```

```python
#import the LinearRegression function from sklearn.linear_model python package
#LinearRegression is a function used to create the Simple Linear regression model
from sklearn.linear_model import LinearRegression
#Assign the LinearRegression function to the regressor variable
regressor=LinearRegression()
#fit method is used to create a model. It will get the input and output/to predict output and create a model
regressor.fit(X_train,Y_train)
```

```python
#coef_ method used to get the weight of the model
weight=regressor.coef_
weight
```

```python
#intercept_ function is used to get the bias of the model
bias=regressor.intercept_
bias
```

```python
#predict function will take input dataset and return the prediction
Y_predict=regressor.predict(X_test)
Y_predict
```

```python
#import r2_score function from sklearn.metrics puthon library package
#r2_score will be used to validate the model
from sklearn.metrics import r2_score
#r2_score will compare the Output/to predict values from the original file with the output/prediction by the model and return the value between o and 1
#if r2_score value is near to 1, it is a good prediction
#if r2_score value is near to 0, it is a bad prediction
r_score=r2_score(Y_test,Y_predict)
r_score
```

```python
#pickle is a python library used to save the model in separate file
import pickle
filename="simple_linear_regression_model.sav"
#dump function in pickle is used to save the file. It will get model(regressor) and filedetails with writebyte as a input and save the model in the file
pickle.dump(regressor,open(filename,'wb'))
```

```python
#load function in pickel is used to load the model. It will get filedetails with readbyte as a input and store the details in variable.
loaded_model=pickle.load(open(filename,'rb'))
#predict function can be used to predict the values from the model It will take values for which we need prediction.
result=loaded_model.predict([[15]])
result
```
